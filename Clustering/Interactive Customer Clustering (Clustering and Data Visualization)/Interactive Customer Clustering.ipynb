```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
```

```python
# Veriyi Yüklüyoruz
```

```python
df = pd.read_csv("social.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Kümeleme için kullanılacak özellikler
```

```python
X = df[['Age', 'EstimatedSalary']]
```

```python
# Ölçeklendirme (Clustering için kritik)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
# K-Means Modeli
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
```

```python
kmeans.fit(X_scaled)
```

```python
# Kümeleri veriye ekliyoruz
```

```python
df['Cluster'] = kmeans.labels_
```

```python
# Save
```

```python
import joblib
```

```python
joblib.dump(kmeans, "kmeans_model.pkl")
```

```python
joblib.dump(scaler, "cluster_scaler.pkl")
```

```python
df.to_csv("social_clustered.csv", index=False)
```
