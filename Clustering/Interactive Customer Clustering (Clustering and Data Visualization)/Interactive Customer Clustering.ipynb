```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import joblib

# 1. Veriyi Yükle
df = pd.read_csv("social.csv")

# Kümeleme için kullanılacak özellikler
X = df[['Age', 'EstimatedSalary']]

# 2. Ölçeklendirme (Clustering için kritik)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. K-Means Modeli (K=3 olarak belirleyelim: Düşük, Orta, Yüksek Gelir/Yaş grubu gibi)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_scaled)

# Kümeleri veriye ekle
df['Cluster'] = kmeans.labels_

# 4. Kaydet
joblib.dump(kmeans, "kmeans_model.pkl")
joblib.dump(scaler, "cluster_scaler.pkl")
df.to_csv("social_clustered.csv", index=False)

print("Model eğitildi ve kümeler oluşturuldu.")
```
