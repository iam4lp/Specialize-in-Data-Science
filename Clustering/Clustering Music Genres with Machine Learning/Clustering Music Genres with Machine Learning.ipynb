Clustering Music Genres with Machine Learning uygulamamız için modelimizi kaydedeceğiz

```python
import pandas as pd
```

```python
# Veriyi Yüklüyoruz
```

```python
df = pd.read_csv("Spotify-2000.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Veri Temizleme (Length sütunundaki virgülleri kaldırma)
```

```python
if df['Length (Duration)'].dtype == 'object':
    df['Length (Duration)'] = df['Length (Duration)'].astype(str).str.replace(',', '').astype(int)
```

```python
# Kullanılacak Ses Özellikleri
```

```python
features = [
    'Beats Per Minute (BPM)', 'Energy', 'Danceability', 
    'Loudness (dB)', 'Liveness', 'Valence', 
    'Length (Duration)', 'Acousticness', 'Speechiness', 'Popularity'
]
```

```python
X = df[features]
```

```python
# Ölçeklendirme (Standard Scaling)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
# K-Means Kümeleme
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
```

```python
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Görselleştirme İçin Boyut İndirgeme (PCA - 2 Boyut)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
principal_components = pca.fit_transform(X_scaled)
```

```python
df['PCA1'] = principal_components[:, 0]
```

```python
df['PCA2'] = principal_components[:, 1]
```

```python
# Save
```

```python
df.to_csv("spotify_clustered.csv", index=False)
```

```python
import joblib
```

```python
joblib.dump(kmeans, "spotify_kmeans.pkl")
```

```python
joblib.dump(scaler, "spotify_scaler.pkl")
```
